```python
import json
from oandapyV20 import API
import oandapyV20.endpoints.instruments as instruments
import pandas as pd

with open('demo_account.json') as f:
    account_keys = json.load(f)

account_keys.keys()
```

```python
api_key = account_keys['token']
api = API(access_token=api_key, environment="practice")
```

```python
instrument = "EUR_USD"
params = {
    "count": 5000,  # Number of data points, maximum is 5000
    "granularity": "H1"  # Time interval, e.g., H1 for hourly data
}

historical_data = instruments.InstrumentsCandles(instrument=instrument, params=params)
d_data = api.request(historical_data)
```

```python
def _flatten_bar_dict(d_bar, price='mid'):
    d_bar = d_bar.copy()
    d_ohlc = d_bar.pop(price)
    return {**d_bar, **d_ohlc}

df_data = pd.DataFrame([_flatten_bar_dict(d_bar) for d_bar in d_data['candles']])
df_data['time'] = pd.to_datetime(df_data['time'])
df_data = df_data[['time', 'o', 'h', 'l', 'c', 'volume', 'complete']]

df_data
```
